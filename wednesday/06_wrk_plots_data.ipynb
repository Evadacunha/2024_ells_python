## Show outliners of ProductionCost


```python
Q1 = pmd["ProductionCost"].quantile(0.25) #first 25% of values
print(Q1)
Q3 = pmd["ProductionCost"].quantile(0.75) #last 25% of values
print(Q3)
IQR = Q3 - Q1 #inter quartile range
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = pmd[(pmd["ProductionCost"] < lower_bound) | (pmd["ProductionCost"] > upper_bound)]
outliers #no outliers exist
```
